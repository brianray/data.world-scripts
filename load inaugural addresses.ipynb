```python

import requests
import zipfile

url = "https://raw.githubusercontent.com/nltk/nltk_data/gh-pages/packages/corpora/inaugural.zip"
r = requests.get(url)
with open("inaugural.zip", "wb") as code:
    code.write(r.content)

zip_ref = zipfile.ZipFile("inaugural.zip", 'r')
zip_ref.extractall("inaugural")
zip_ref.close()
```
