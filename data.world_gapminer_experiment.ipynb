```python
import re
from lxml import html
data = open("Data.htm", "r").read() # Data.htm from https://www.gapminder.org/data/
tree = html.fromstring(data)
titles = [x.text for x in tree.xpath('//*[@id="indicators-table"]/tbody/tr[*]/td[1]/a')]
sources = [x.text for x in tree.xpath('//*[@id="indicators-table"]/tbody/tr[*]/td[2]/a')]
tags = [[str(x.text).lower(), 'gapminder'] for x in tree.xpath('//*[@id="indicators-table"]/tbody/tr[*]/td[3]')]
links = [link.values()[1] for link in tree.xpath('//*[@id="indicators-table"]/tbody/tr[*]/td[5]/a[1]')]


data = tuple(zip(titles, links, sources, tags))
```

```python
import os.path
import urllib.request
import datadotworld as dw
import re

client = dw.api_client()

for title, link, based, tags in data:
    orglink = link
    link = link.strip().replace("&amp;", "&").replace("xls", "csv")
    title = re.sub(r'\W+', ' ', title)
    filename = "{}.csv".format(title.strip())
    title = "gapminder {}".format(title)[:30]
    data_args = dict(owner_id="brianray",
                     title=title,
                     description="{} based on {}".format(title, based)[:120],
                     tags=tags,
                     license='CC-BY',
                     visibility="OPEN",
                     files={filename: link})
    print(data_args)
    try:
        client.create_dataset(**data_args)
    except Exception as e:
        print(e)
        continue
    print("---")
    
    
    
```
